```python
import pandas as pd
#Read the three files into python as dataframes
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
#Show the DataFrame's shape
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
#Standardize header names
file1.columns
file1 = file1.rename(columns={"ST": "State", "GENDER": "Gender", 'Customer Lifetime Value': 'Customer_Lifetime_Value', 'Monthly Premium Auto': 'Monthly_Premium_Auto', 'Number of Open Complaints': 'Number_of_Open_Complaints', 'Total Claim Amount': 'Total_Claim_Amount', 'Policy Type': 'Policy_Type', 'Vehicle Class': 'Vehicle_Class'})
file1.columns
```

```python
file2.columns
file2 = file2.rename(columns={"ST": "State", "GENDER": "Gender", 'Customer Lifetime Value': 'Customer_Lifetime_Value', 'Monthly Premium Auto': 'Monthly_Premium_Auto', 'Number of Open Complaints': 'Number_of_Open_Complaints', 'Total Claim Amount': 'Total_Claim_Amount', 'Policy Type': 'Policy_Type', 'Vehicle Class': 'Vehicle_Class'})
file2.columns
```

```python
file3.columns
file3 = file3.rename(columns={'Customer Lifetime Value': 'Customer_Lifetime_Value', 'Monthly Premium Auto': 'Monthly_Premium_Auto', 'Number of Open Complaints': 'Number_of_Open_Complaints', 'Total Claim Amount': 'Total_Claim_Amount', 'Policy Type': 'Policy_Type', 'Vehicle Class': 'Vehicle_Class'})
file3.columns
```

```python
#Rearrange the columns in the dataframe as needed
file1_order = file1.loc[:, ['Customer', 'State', 'Education', 'Gender', 'Income', 'Customer_Lifetime_Value', 'Monthly_Premium_Auto', 'Policy_Type', 'Vehicle_Class', 'Number_of_Open_Complaints',
       'Total_Claim_Amount']]
```

```python
column_names = file1_order.columns
data = pd.DataFrame(columns = column_names)
```

```python
#Concatenate the three dataframes
data = pd.concat([data,file1,file2,file3], axis = 0)
```

```python
data = data.reset_index()
```

```python
data
```

```python
#Which columns are numerical?
data_numerical = data.select_dtypes(exclude='object')
```

```python
data_numerical
```

```python
#Which columns are categorical?
data_categorical = data.select_dtypes(include='object')
```

```python
data_categorical
```

```python
#Delete the column education and the number of open complaints from the dataframe
data.drop(['Education', 'Number_of_Open_Complaints'], axis = 1, inplace=True)
```

```python
data
```

```python
#Correct the values in the column customer lifetime value 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type
data["Customer_Lifetime_Value"].fillna(0, inplace = True)
```

```python
data["Customer_Lifetime_Value"] = data["Customer_Lifetime_Value"].str.rstrip('%').astype(float) / 100
```

```python
data["Customer_Lifetime_Value"]=pd.to_numeric(round(data["Customer_Lifetime_Value"],1))
```

```python
data
```

```python
#Check for duplicate rows in the data and remove if any
data.drop_duplicates(inplace=True)
```

```python
#Filter out the data for customers who have an income of 0 or less
data_filtered = data['Customer'][data['Income']<=0]
```
